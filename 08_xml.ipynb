## More on XML parsing

Here's the sample file in the data folder ("data/menu.xml")

```
<?xml version="1.0" encoding="UTF-8"?>
<breakfast_menu>
	<food>
		<name>Belgian Waffles</name>
		<price>$5.95</price>
		<description>Two of our famous Belgian Waffles with plenty of real maple syrup
		</description>
		<calories>650</calories>
	</food>
	<food>
		<name>Strawberry Belgian Waffles</name>
		<price>$7.95</price>
		<description>Light Belgian waffles covered with strawberries and whipped cream</description>
		<calories>900</calories>
	</food>
	<food>
		<name>Berry-Berry Belgian Waffles</name>
		<price>$8.95</price>
		<description>Light Belgian waffles covered with an assortment of fresh berries and
                whipped cream</description>
		<calories>900</calories>
	</food>
	<food>
		<name>French Toast</name>
		<price>$4.50</price>
		<description>Thick slices made from our homemade sourdough bread</description>
		<calories>600</calories>
	</food>
	<food>
		<name>Homestyle Breakfast</name>
		<price>$6.95</price>
		<description>Two eggs, bacon or sausage, toast, and our ever-popular hash browns</description>
		<calories>950</calories>
	</food>
</breakfast_menu>
```

````python
import xml.etree.ElementTree as ET     # Define 'ET' as the shorthand for 'xml.etree.ElementTree'

fin = open("data/menu.xml")
data = fin.read()

tree = ET.fromstring(data)             # ask ET (ElementTree) to read the data and create a 'tree'.
print("tag = {}".format(tree.tag.strip()))
print("text = {}".format(tree.text.strip()))
print("-----")
for f in tree.findall('food'):         # find all the <food> tags.
    for element in f.iter():           # iterate all the tags, under the food tag (including itself)
        print("tag = {}".format(element.tag.strip()))     # get the tag name
        print("text = {}".format(element.text.strip()))   # get the text value
    print("-----")    
````
