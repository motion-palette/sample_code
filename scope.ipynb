```python
x = 10     # global
y = 20     # global
z = 30     # global

def add_three_nums(a, b, c):     # 'a', 'b', and 'c' are arguments. Arguments are local to the function.
    total = a + b + c          # 'total' is a local variable within 'add_two_nums()'
    return total

a = add_three_nums(x, y, z)
print(a)
```

```python
x = 10     # global
y = 20     # global

def a_func(n, m):
    k = 30
    def add_two_nums_and_ten(a, b):            # this function is local to 'a_func()'
        total = a + b + k                      # 'k' is accessible
        return total                           # total is local to 'add_three_num()'

    print(k)           # 'k' is in the current scope.
    # print(total)     # 'total' is outside of the current scope.

    result = 0
    for i in range(5):
        result += add_two_nums_and_ten(n, m)

    return result

a = a_func(x, y)     # ONLY 'a', 'x' and 'y' are in the current scope.
print(a)
```
